<p>It would be interesting to find out the distributuion of loan amounts.</p>
<p> Given that laon amounts is a continous variable determining the minimum and maximum values can help split the loan amounts into bins. </p>

```python
def max_and_min(column):
    maximum = column.max()
    minimum = column.min()
    return maximum,minimum
max_current, min_current = max_and_min(current_loans.loan_amnt)
max_fully, min_fully = max_and_min(fully_paid_loans.loan_amnt)
max_charged, min_charged = max_and_min(charged_off_loans.loan_amnt)
max_and_min_list = [('Current Loans',max_current, min_current),
                    ('Fully Paid Loans',max_fully, min_fully),
                   ('Charged off Loans',max_charged, min_charged)]
max_min_header = ['Dataset','Max Value','Min Value']
print(tabulate(max_and_min_list,headers = max_min_header))
```

<p>On the three dataset the overall minimum loan amount given as a loan is 500 and the maximum loan amount is 35,000</p>

```python
def amount_array(column):
    list_array = np.array(column.loan_amnt)
    return list_array
column_current_loan = np.array(current_loans.loan_amnt)
column_fully_loan = np.array(fully_paid_loans.loan_amnt)
column_charged_loan = np.array(charged_off_loans.loan_amnt)

def num_count(column_array):
    A = 0
    B = 0
    C = 0
    D = 0
    E = 0
    F = 0
    G = 0
    for i in column_array:
        if i >= 0 and i <= 4999:
            A = A + 1
        elif i >= 5000 and i <= 9999:
            B = B + 1
        elif i >= 10000 and i <= 14999:
            C = C + 1
        elif i >= 15000 and i <= 19999:
            D = D + 1
        elif i >= 20000 and i <= 24999:
            E = E + 1
        elif i >= 25000 and i <= 29999:
            F = F + 1 
        else:
            G = G + 1
    value_seg = [A,B,C,D,E,F,G]
    return value_seg

current_loan_bin = num_count(column_current_loan )
fully_paid_bin = num_count(column_fully_loan)
charged_off_bin = num_count(column_charged_loan)

setter = ['0 - 4999','5000 - 9999','10000 - 14999','15000 - 19999','20000 - 24999','25000 - 29999','30000 - 35000']
labels = ['A','B','C','D','E','F','G']

for i in range(len(setter)):
    print('{}   :   {}'.format(labels[i],setter[i]))
    
fig, ((ax1,ax2,ax3)) = plt.subplots(1, 3,figsize=(15, 6))
sizes_1 = current_loan_bin
sizes_2 = fully_paid_bin
sizes_3 = charged_off_bin

ax1.pie(sizes_1,autopct='%1.1f%%',labels = labels, shadow=True, startangle=140)
ax1.set_title('Current Loans')
ax2.pie(sizes_2,autopct='%1.1f%%',labels = labels, shadow=True, startangle=140)
ax2.set_title('Fully Paid Loans')
ax3.pie(sizes_3,autopct='%1.1f%%',labels = labels, shadow=True, startangle=140)
ax3.set_title('Charged Off Loans')
plt.axis('equal')
plt.show()
```

It would be important to see the distribution for the purposes that people took the loans 

```python
purpose_loans_current_unique = np.array(current_loans.purpose.unique())
print('There are {} unique different reasons for why people take on loans.'.format(len(purpose_loans_unique)))
purpose_loans_current = np.array(current_loans.purpose)
purpose_loans_fully = np.array(fully_paid_loans.purpose)
purpose_loans_charged = np.array(charged_off_loans.purpose)
purpose_current_dict = {}
purpose_fully_dict = {}
purpose_charged_dict = {}
def update_dict_purpose(column,dictionary):
    for i in purpose_loans_unique:
        dictionary[i] = 0
    for i in column:
        for j in dictionary.keys():
            if i == j:
                dictionary[j] +=1
purpose_current_values  = []
purpose_fully_values = []
purpose_charged_values = []
def create_list_purpose(dictionary,list_values):
    for j in dictionary.keys():
        list_values.append(dictionary[j])
update_dict_purpose(purpose_loans_current,purpose_current_dict)
update_dict_purpose(purpose_loans_fully,purpose_fully_dict)
update_dict_purpose(purpose_loans_charged,purpose_charged_dict)
create_list_purpose(purpose_current_dict,purpose_current_values)
create_list_purpose(purpose_fully_dict,purpose_fully_values)
create_list_purpose(purpose_charged_dict,purpose_charged_values)
```

```python
fig, ((ax1,ax2,ax3)) = plt.subplots(1, 3,figsize=(20, 6))
ax1.barh(purpose_loans_unique,purpose_current_values) 
ax1.set_title('Current Loans')
ax2.barh(purpose_loans_unique,purpose_fully_values)
ax2.set_title('Fully Paid Loans')
ax3.barh(purpose_loans_unique,purpose_charged_values)
ax3.set_title('Charged Off Loans')
for ax in fig.get_axes():
    ax.label_outer()
for i in (ax1,ax2,ax3):
    i.set(xlabel='NUMBER OF PEOPLE')
ax1.set(ylabel='PURPOSES')
```
